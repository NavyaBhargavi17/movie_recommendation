```python
import pandas as pd

movies = pd.read_csv(
    '/content/sample_data/tmdb_5000_movies.csv',
    engine='python',
    encoding='utf-8',
    on_bad_lines='skip'
)

credits = pd.read_csv(
    '/content/sample_data/tmdb_5000_credits.csv',
    engine='python',
    encoding='utf-8',
    on_bad_lines='skip'
)

movies.head()
```

```python
credits.head()
```

```python
movies = movies[['movie_id', 'title', 'tags']]
movies.head()

```

```python
movies.columns
```

```python
movies = movies.rename(columns={'id': 'movie_id'})
```

```python
import ast
```

```python
import ast

def safe_convert(text):
    try:
        data = ast.literal_eval(text)
        return [i['name'] for i in data]
    except:
        return []

```

```python
def safe_fetch_director(text):
    try:
        data = ast.literal_eval(text)
        for i in data:
            if i.get('job') == 'Director':
                return i.get('name')
        return ''
    except:
        return ''
```

```python
movies['genres'] = movies['genres'].apply(safe_convert)
movies['keywords'] = movies['keywords'].apply(safe_convert)
```

```python
credits['cast'] = credits['cast'].apply(safe_convert)
credits['crew'] = credits['crew'].apply(safe_fetch_director)
```

```python
movies[['genres', 'keywords']].head()
credits[['cast', 'crew']].head()
```

```python
movies = movies.merge(
    credits[['movie_id', 'cast', 'crew']],
    on='movie_id'
)

movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew'].apply(lambda x: [x])
)
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x).lower())
```

```python
movies = movies[['movie_id', 'title', 'tags']]
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = tfidf.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()

    if movie_name not in movies['title'].str.lower().values:
        return "Movie not found"

    index = movies[movies['title'].str.lower() == movie_name].index[0]

    distances = similarity[index]
    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    return [movies.iloc[i[0]].title for i in movie_list]
```

```python
recommend("Avatar")
```

```python
recommend("Inception")
recommend("Batman Begins")
```

```python
import pickle

pickle.dump(movies, open("movies.pkl", "wb"))
pickle.dump(similarity, open("similarity.pkl", "wb"))
```

```python
%%writefile app.py
import streamlit as st
import pickle

st.set_page_config(page_title="Movie Recommendation System")

st.title("🎬 Movie Recommendation System")
st.write("Select a movie to get recommendations")

movies = pickle.load(open("movies.pkl", "rb"))
similarity = pickle.load(open("similarity.pkl", "rb"))

def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = similarity[index]
    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]
    return [movies.iloc[i[0]].title for i in movie_list]

selected_movie = st.selectbox(
    "Choose a movie",
    movies['title'].values
)

if st.button("Recommend"):
    for movie in recommend(selected_movie):
        st.write("🎥", movie)
```
